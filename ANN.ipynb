```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Customer_info.csv")
df
```

```python
df.info()
```

# Feature Selection

```python
# we are excluding RowNumber , CustomerId, and Surname as they are not needed.
x = df.iloc[:, 3:-1]
y = df.iloc[:,13]
```

```python
x.head(10)
```

```python
print(x['Geography'].unique())
```

# Creating dummies variable for Geography and Gender

```python
Geog = pd.get_dummies(x['Geography'],drop_first=True)
Gen = pd.get_dummies(x['Gender'],drop_first=True)

```

```python
# adding the dummies column to x
x = pd.concat([x,Geog,Gen], axis=1)
```

```python
#dropping the catagorical features
x = x.drop(["Geography","Gender"], axis=1)
```

```python
x.head()
```

```python
#splitting the data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# feature scaling to bring features into same scale range.
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

# Implementing ANN(Artificial Nural Network)

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential  #
from keras.layers import Dense       # for creating hidden leayers
from keras.layers import LeakyReLU,PReLU,ELU #
from keras.layers import Dropout # using for regularization
```

```python
model = Sequential()

#first hidden layer
model.add(Dense(10, activation='relu', input_dim=11)) # input dim = number of features 

#second hidden layer
model.add(Dense(12, activation='relu'))


#second hidden layer
model.add(Dense(15, activation='relu'))

#Output layer
model.add(Dense(1,activation='sigmoid'))

# Havnt used drop out becuse we dont have deep dense layers. In our case it might reduce the accuracy
```

```python
model.summary()
```

# Compiling the ANN ( using optimizer and Loss function)

```python
model.compile(optimizer="Adam", loss="binary_crossentropy", metrics=['accuracy'])
```

```python
# Training the data
model_train = model.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

```python
print(model_train.history.keys())
```

# Predicting test Result

```python
y_pred = model.predict(X_test)

# converting the 0 and 1 into true or fale 
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
score
```
